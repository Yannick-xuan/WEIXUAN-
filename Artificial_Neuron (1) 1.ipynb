```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulating artificial neuron data
num_samples = 5000
time = np.linspace(0, 10, num_samples)  # Time array

# Generating membrane potential using a sinusoidal function
frequency = 0.5  # Frequency of membrane potential oscillation
amplitude = 20  # Amplitude of membrane potential oscillation
membrane_potential = amplitude * np.sin(2 * np.pi * frequency * time)

# Creating a DataFrame to store the artificial neuron data
neuron_data = {
    'Time': time,
    'Membrane_Potential': membrane_potential
}

df_neuron = pd.DataFrame(neuron_data)

# Plotting the simulated artificial neuron data
plt.figure(figsize=(8, 4))
plt.plot(df_neuron['Time'], df_neuron['Membrane_Potential'], label='Membrane Potential')
plt.xlabel('Time')
plt.ylabel('Membrane Potential')
plt.title('Artificial Neuron Data')
plt.legend()
plt.grid(True)
plt.show()

# Save the artificial neuron data to an Excel file
excel_filename = 'artificial_neuron_data.xlsx'
df_neuron.to_excel(excel_filename, index=False)

print(f"Artificial neuron data saved to {excel_filename}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load artificial neuron data
# Assuming df_neuron contains the artificial neuron data (time and membrane potential)
# Replace df_neuron with the actual DataFrame containing your data
time = df_neuron['Time'].values
neuron_membrane_potential = df_neuron['Membrane_Potential'].values

# Constants for LIF model
R = 10.0      # Membrane resistance in megaohms
C = 1.0       # Membrane capacitance in microfarads
V_rest = -70  # Resting membrane potential in mV
V_thresh = -55  # Threshold potential in mV
V_reset = -75  # Reset potential in mV
I_inj = 2.0   # Injected current in microamperes

# Simulation parameters for LIF model
dt = 0.01    # Time step in milliseconds
num_steps = len(time)  # Number of steps based on artificial neuron data

# Initialize variables for LIF model
V = np.zeros(num_steps)
spike_count = 0

# Initial condition using artificial neuron data
V[0] = neuron_membrane_potential[0]  # Set initial membrane potential using artificial neuron data

# Leaky Integrate-and-Fire equations simulation using artificial neuron data
for i in range(1, num_steps):
    dVdt = (-(V[i - 1] - V_rest) + R * I_inj) / (R * C)
    V[i] = V[i - 1] + dt * dVdt

    # If membrane potential reaches threshold, reset and count spike
    if V[i] >= V_thresh:
        V[i] = V_reset
        spike_count += 1

# Plotting the LIF model simulation with artificial neuron data
plt.figure(figsize=(8, 4))
plt.plot(time, V, label='Membrane Potential (LIF Model)')
plt.xlabel('Time')
plt.ylabel('Membrane Potential (mV)')
plt.title('Leaky Integrate-and-Fire Model Simulation with Artificial Neuron Data')
plt.legend()
plt.grid(True)
plt.show()

# Print spike count
print(f"Total spikes generated: {spike_count}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants for the Integrate-and-Fire neuron model
R = 10.0      # Membrane resistance in megaohms
C = 1.0       # Membrane capacitance in microfarads
V_rest = -70  # Resting membrane potential in mV
V_thresh = -55  # Threshold potential in mV
V_reset = -75  # Reset potential in mV
I_inj = 2.0   # Injected current in microamperes

# Simulation parameters
dt = 0.01    # Time step in milliseconds
t = np.arange(0, 50, dt)  # Time array
num_steps = len(t)

# Initialize variables
V = np.zeros(num_steps)

# Initial condition
V[0] = V_rest  # Initial membrane potential

# Simulate the behavior of the Integrate-and-Fire neuron model
for i in range(1, num_steps):
    dVdt = (-(V[i - 1] - V_rest) + R * I_inj) / (R * C)
    V[i] = V[i - 1] + dt * dVdt

    # If membrane potential reaches threshold, reset
    if V[i] >= V_thresh:
        V[i] = V_reset

# Plotting the membrane potential over time
plt.figure(figsize=(10, 5))
plt.plot(t, V, label='Membrane Potential')
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Potential (mV)')
plt.title('Integrate-and-Fire Neuron Model Simulation')
plt.legend()
plt.grid(True)
plt.show()
```

# Simulating the Hodgkin-Huxley (HH) Model

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants for the HH model
C_m = 1.0   # Membrane capacitance in microfarads/cm^2
g_Na = 120.0  # Sodium (Na) maximum conductances in mS/cm^2
g_K = 36.0    # Potassium (K) maximum conductances in mS/cm^2
g_L = 0.3     # Leak maximum conductances in mS/cm^2
E_Na = 50.0   # Sodium (Na) reversal potentials in mV
E_K = -77.0   # Potassium (K) reversal potentials in mV
E_L = -54.387  # Leak reversal potentials in mV

# Simulation parameters
dt = 0.01     # Time step in milliseconds
t = np.arange(0, 50, dt)  # Time array
num_steps = len(t)

# Initialize variables
V = np.zeros(num_steps)
m = np.zeros(num_steps)
h = np.zeros(num_steps)
n = np.zeros(num_steps)

# Initial conditions
V[0] = -65.0  # Initial membrane potential in mV
m[0] = 0.05
h[0] = 0.6
n[0] = 0.32

# Hodgkin-Huxley equations
for i in range(1, num_steps):
    alpha_m = (0.1 * (V[i - 1] + 40)) / (1 - np.exp(-(V[i - 1] + 40) / 10))
    beta_m = 4 * np.exp(-(V[i - 1] + 65) / 18)
    alpha_h = 0.07 * np.exp(-(V[i - 1] + 65) / 20)
    beta_h = 1 / (1 + np.exp(-(V[i - 1] + 35) / 10))
    alpha_n = (0.01 * (V[i - 1] + 55)) / (1 - np.exp(-(V[i - 1] + 55) / 10))
    beta_n = 0.125 * np.exp(-(V[i - 1] + 65) / 80)

    m[i] = m[i - 1] + dt * (alpha_m * (1 - m[i - 1]) - beta_m * m[i - 1])
    h[i] = h[i - 1] + dt * (alpha_h * (1 - h[i - 1]) - beta_h * h[i - 1])
    n[i] = n[i - 1] + dt * (alpha_n * (1 - n[i - 1]) - beta_n * n[i - 1])

    g_Na_t = g_Na * (m[i] ** 3) * h[i]
    g_K_t = g_K * (n[i] ** 4)
    g_L_t = g_L

    I_Na = g_Na_t * (V[i - 1] - E_Na)
    I_K = g_K_t * (V[i - 1] - E_K)
    I_L = g_L_t * (V[i - 1] - E_L)

    dVdt = (-(I_Na + I_K + I_L) + 10) / C_m
    V[i] = V[i - 1] + dt * dVdt

# Plotting results for HH model
plt.figure(figsize=(10, 5))
plt.plot(t, V, label='Membrane Potential (HH Model)')
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Potential (mV)')
plt.title('Hodgkin-Huxley Model Simulation')
plt.legend()
plt.grid(True)
plt.show()
```

# Simulating the Leaky Integrate-and-Fire (LIF) Model

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants for the LIF model
R = 10.0      # Membrane resistance in megaohms
C = 1.0       # Membrane capacitance in microfarads
V_rest = -70  # Resting membrane potential in mV
V_thresh = -55  # Threshold potential in mV
V_reset = -75  # Reset potential in mV
I_inj = 2.0   # Injected current in microamperes

# Simulation parameters
dt = 0.01    # Time step in milliseconds
t = np.arange(0, 50, dt)  # Time array
num_steps = len(t)

# Initialize variables
V = np.zeros(num_steps)
spike_count = 0

# Initial condition
V[0] = V_rest  # Initial membrane potential

# Leaky Integrate-and-Fire equations
for i in range(1, num_steps):
    dVdt = (-(V[i - 1] - V_rest) + R * I_inj) / (R * C)
    V[i] = V[i - 1] + dt * dVdt

    # If membrane potential reaches threshold, reset and count spike
    if V[i] >= V_thresh:
        V[i] = V_reset
        spike_count += 1

# Plotting results for LIF model
plt.figure(figsize=(10, 5))
plt.plot(t, V, label='Membrane Potential (LIF Model)')
plt.xlabel('Time (ms)')
plt.ylabel('Membrane Potential (mV)')
plt.title('Leaky Integrate-and-Fire Model Simulation')
plt.legend()
plt.grid(True)
plt.show()

# Print spike count
print(f"Total spikes generated: {spike_count}")
```
