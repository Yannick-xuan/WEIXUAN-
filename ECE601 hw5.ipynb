```python
# import required libraries
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

# Load and prepare the CIFAR-10 dataset
(cifar_images, cifar_labels), (x_test, y_test) = datasets.cifar10.load_data()

# Normalize the input images
cifar_images = cifar_images / 255.0
x_test = x_test / 255.0

# Convert labels to one-hot encoding
cifar_labels = tf.keras.utils.to_categorical(cifar_labels, num_classes=10)
y_test = tf.keras.utils.to_categorical(y_test, num_classes=10)


# Design four CNN models to classify CIFAR-10 images with the following features

'''
\begin{itemize}
    --- First structure includes two convolutional layers with ReLU activation.
    --- The first layer with 16 filters and the second filter 32 filters. Use the filter size (5,5)
    --- Second structure includes two convolutional layers with ReLU activation.
    --- The first layer with 32 filters and the second filter 64 filters. Use the filter size (3,3)
    --- Third structure includes three convolutional layers with ReLU activation.
    --- The first layer with 16 filters, the second layer 32 filters, and the third filter 64 filters. Use the filter size (3,3)
    --- Forth structure includes three convolutional layers with ReLU activation.
    --- The first layer with 32 filters, the second layer 64 filters, and the third filter 64 filters. Use the filter size (3,3)
    \item Pooling layers following each convolutional layer to reduce the spatial dimensions of the feature maps.
    \item A flattening layer to convert the 2D feature maps into a 1D vector.
    \item Dense layers with 64 nodes and a relu activation function.
    \item Dense layers that lead to the final classification output. The last layer should have 10 units with a softmax activation function, corresponding to the CIFAR-10 classes.
\end{itemize}
'''

'''
Use an optimizer and loss function suitable for a classification problem
(categorical cross-entropy is recommended), and include accuracy as a metric.
'''

pass
```

```python
# First structure
cnn_model_1 = models.Sequential([
    layers.Conv2D(16, (5, 5), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(32, (5, 5), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model
cnn_model_1.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
# Train the model for at least 10 epochs.
cnn_history_1 = cnn_model_1.fit(cifar_images[..., tf.newaxis], cifar_labels, epochs=10,
                                validation_data=(x_test[..., tf.newaxis], y_test))
# Evaluate the model
test_loss_1, test_acc_1 = cnn_model_1.evaluate(x_test[..., tf.newaxis],  y_test, verbose=0)
print('\nTest accuracy:', test_acc_1)
```

```python
# Second structure
cnn_model_2 = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model
cnn_model_2.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
# Train the model for at least 10 epochs.
cnn_history_2 = cnn_model_2.fit(cifar_images[..., tf.newaxis], cifar_labels, epochs=10,
                                validation_data=(x_test[..., tf.newaxis], y_test))
# Evaluate the model
test_loss_2, test_acc_2 = cnn_model_2.evaluate(x_test[..., tf.newaxis],  y_test, verbose=2)
print('\nTest accuracy:', test_acc_2)
```

```python
# Third structure
cnn_model_3 = models.Sequential([
    layers.Conv2D(16, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model
cnn_model_3.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
# Train the model for at least 10 epochs.
cnn_history_3 = cnn_model_3.fit(cifar_images[..., tf.newaxis], cifar_labels, epochs=10,
                                validation_data=(x_test[..., tf.newaxis], y_test))
# Evaluate the model
test_loss_3, test_acc_3 = cnn_model_3.evaluate(x_test[..., tf.newaxis],  y_test, verbose=2)
print('\nTest accuracy:', test_acc_3)
```

```python
# Third structure
cnn_model_4 = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile the model
cnn_model_4.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
# Train the model for at least 10 epochs.
cnn_history_4 = cnn_model_4.fit(cifar_images[..., tf.newaxis], cifar_labels, epochs=10,
                                validation_data=(x_test[..., tf.newaxis], y_test))
# Evaluate the model
test_loss_4, test_acc_4 = cnn_model_4.evaluate(x_test[..., tf.newaxis],  y_test, verbose=2)
print('\nTest accuracy:', test_acc_4)
```

```python
# Plot training & validation accuracy values for all different models (In total,  )
plt.plot(cnn_history_1.history['accuracy'], label='CNN_1 Train')
plt.plot(cnn_history_1.history['val_accuracy'], label='CNN_1 Test')
plt.plot(cnn_history_2.history['accuracy'], label='CNN_2 Train')
plt.plot(cnn_history_2.history['val_accuracy'], label='CNN_2 Test')
plt.plot(cnn_history_3.history['accuracy'], label='CNN_3 Train')
plt.plot(cnn_history_3.history['val_accuracy'], label='CNN_3 Test')
plt.plot(cnn_history_4.history['accuracy'], label='CNN_4 Train')
plt.plot(cnn_history_4.history['val_accuracy'], label='CNN_4 Test')
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(cnn_history_1.history['loss'], label='CNN_1 Train')
plt.plot(cnn_history_1.history['val_loss'], label='CNN_1 Test')
plt.plot(cnn_history_2.history['loss'], label='CNN_2 Train')
plt.plot(cnn_history_2.history['val_loss'], label='CNN_2 Test')
plt.plot(cnn_history_3.history['loss'], label='CNN_3 Train')
plt.plot(cnn_history_3.history['val_loss'], label='CNN_3 Test')
plt.plot(cnn_history_4.history['loss'], label='CNN_4 Train')
plt.plot(cnn_history_4.history['val_loss'], label='CNN_4 Test')
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(loc='upper left')
plt.show()
```
